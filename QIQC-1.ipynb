### 量子比特的数学抽象
量子比特：$$ \left\vert \psi \right\rangle = \alpha \left\vert 0 \right\rangle + \beta \left\vert 1 \right\rangle$$  
例如, (1)+态：
$$ \left\vert + \right\rangle = \frac{1}{2} \left(\left\vert 0 \right\rangle + \beta \left\vert 1 \right\rangle \right)$$ 
(2)-态：
$$ \left\vert - \right\rangle = \frac{1}{2} \left(\left\vert 0 \right\rangle - \left\vert 1 \right\rangle \right)$$  

### 量子比特的角度表示

在$ \left\vert \psi \right\rangle = \alpha \left\vert 0 \right\rangle + \beta \left\vert 1 \right\rangle $ 中，有条件 $ \alpha ^2 + \beta ^2 =1 $, 可改用角度表示： 

\begin{aligned}
    \left\vert \psi \right\rangle &=\alpha\left\vert 0 \right\rangle +\beta\left\vert 1 \right\rangle \\
    &=e^{i\gamma} \left(\cos\frac{\theta}{2}\left\vert 0 \right\rangle+e^{i\varphi} \sin\frac{\theta}{2}\left\vert 1 \right\rangle\right) \\
    &=\cos\frac{\theta}{2}\left\vert 0 \right\rangle + e^{i\varphi} \sin\frac{\theta}{2}\left\vert 1 \right\rangle
  \end{aligned}

改 $(\alpha,\beta)$ 为 $(\theta,\varphi)$ 

### Bloch球面 
在（$r,\theta,\varphi$）坐标系，由于$r\equiv 1$, 量子比特可用$(\theta,\varphi)$ 确定的球面描述。

### 量子比特的矩阵表示
量子态与系数矩阵有对应关系    

$$\begin{matrix} 
    \left\vert \psi \right\rangle =\begin{pmatrix}
            \alpha\\
            \beta
    \end{pmatrix},
    &    
    \left\vert 0 \right\rangle 
    =\begin{pmatrix}
        1\\
        0
    \end{pmatrix},
    &
    \left\vert 1 \right\rangle 
    =\begin{pmatrix}
        0\\
        1
    \end{pmatrix}
    \end{matrix}$$



### 量子比特的密度矩阵
$$
\begin{aligned}
    \rho &= \left\vert \psi \right\rangle \left\langle \psi \right\vert  \\ 
    &= \begin{pmatrix}
            \alpha\\
            \beta
    \end{pmatrix} 
    \begin{pmatrix}
            \alpha ^* &
            \beta ^*
    \end{pmatrix} \\
    &=    \begin{pmatrix}
            \alpha \alpha ^* &  \alpha \beta ^* \\
          \alpha \beta &  \beta \beta ^*
    \end{pmatrix} 
\end{aligned} $$   

### 量子比特的泡利展开

\begin{aligned}
   \left\vert \psi \right\rangle \left\langle \psi \right\vert  
    &=
    \begin{pmatrix}
            \alpha \alpha ^* &  \alpha \beta ^* \\
          \alpha \beta &  \beta \beta ^*
   \end{pmatrix} 
 \\
 &=
    \begin{pmatrix}
            \cos\frac{\theta}{2} \cos^* \frac{\theta}{2}  &  \cos\frac{\theta}{2}  (e^{i\varphi} \sin\frac{\theta}{2}) ^* \\
          \cos\frac{\theta}{2}  e^{i\varphi} \sin\frac{\theta}{2} &  e^{i\varphi} \sin\frac{\theta}{2} (e^{i\varphi} \sin\frac{\theta}{2}) ^*
   \end{pmatrix} 
 \\
    &= \frac{1}{2}I +\frac{1}{2}\sin\theta\cos\varphi\sigma_x +\frac{1}{2}\sin\theta\sin\varphi\sigma_y+ \frac{1}{2}\cos\theta\sigma_z 
\end{aligned}

#### 泡利矩阵

$$ \sigma_x =
\begin{pmatrix}
    0 & 1 \\
    1 & 0 
\end{pmatrix}, \qquad
\sigma_y =
\begin{pmatrix}
    0 & -i \\
    i & 0 
\end{pmatrix} $$
$$
\sigma_z =
\begin{pmatrix}
    1 & 0 \\
    0 & -1 
\end{pmatrix} , \qquad
I =
\begin{pmatrix}
    1 & 0 \\
    0 & 1 
\end{pmatrix}
$$

构成 $ C^2$ 空间的正交归一完全集

### 双量子比特


#### 双量子比特的计算基矢
- 两个经典比特有四种状态
  $$ 00，01， 10， 11  $$  
- 两个量子比特有四种计算基矢态
  $$  \left\vert 00 \right\rangle， \left\vert 01 \right\rangle， \left\vert 10 \right\rangle，\left\vert 11 \right\rangle$$

- 矩阵表示是原矩阵的直积 

$$  \left\vert 01 \right\rangle = \left\vert 0 \right\rangle \otimes \left\vert 1 \right\rangle =    
\begin{pmatrix}
    1\\
    0
\end{pmatrix}
\otimes
\begin{pmatrix}
    0\\
    1
\end{pmatrix}
=
\begin{pmatrix}
    1 \otimes \begin{pmatrix}
        0\\
        1
    \end{pmatrix}\\
    0 \otimes \begin{pmatrix}
        0\\
        1
    \end{pmatrix}
\end{pmatrix}
=
\begin{pmatrix}
    0\\
    1\\
    0\\
    0
\end{pmatrix}  $$ 


$$  \left\vert 00 \right\rangle= \begin{pmatrix}
    1\\
    0\\
    0\\
    0
\end{pmatrix}, \left\vert 01 \right\rangle= \begin{pmatrix}
    0\\
    1\\
    0\\
    0
\end{pmatrix}, \left\vert 10 \right\rangle= \begin{pmatrix}
    0\\
    0\\
    1\\
    0
\end{pmatrix}, \left\vert 11 \right\rangle= \begin{pmatrix}
    0\\
    0\\
    0\\
    1
\end{pmatrix}  $$


#### 双量子比特的基矢展开

  $$  \left\vert \Psi \right\rangle = \alpha_{00}\left\vert 00 \right\rangle +  \alpha_{01}\left\vert 01 \right\rangle +  \alpha_{10}\left\vert 10 \right\rangle +  \alpha_{11}\left\vert 11 \right\rangle   $$ 

 归一化条件：
 $$ \sum_{ij=0,1} |\alpha_{ij}|^2= 1 $$

#### 双量子比特的测量

<span class = "exam"> 例-1: </span> 对双量子比特系统的第一个位进行测量，测量结果为$\left\vert 0 \right\rangle$, 求测量后体系的态函数

<span class = "answer"> 解: </span> 设体系的初态为 $$  \left\vert \Psi_0 \right\rangle = \alpha_{00}\left\vert 00 \right\rangle +  \alpha_{01}\left\vert 01 \right\rangle +  \alpha_{10}\left\vert 10 \right\rangle +  \alpha_{11}\left\vert 11 \right\rangle   $$
测量导致第一个量子位坍缩，则有：
$$  
\begin{aligned}
\left\vert \Psi_{f} \right\rangle &= \alpha_{00}\left\vert 00 \right\rangle +  \alpha_{01}\left\vert 01 \right\rangle \\ 
&= \left\vert 0 \right\rangle \otimes (\alpha_{00}\left\vert 0 \right\rangle +  \alpha_{01}\left\vert 1 \right\rangle) 
\end{aligned}$$


### 三量子比特

- 三量子比特有8个计算基矢
  $$ \left\vert 000 \right\rangle, \left\vert 001 \right\rangle, \left\vert 010 \right\rangle, \left\vert 011 \right\rangle $$ 
  $$ \left\vert 100 \right\rangle, \left\vert 101 \right\rangle, \left\vert 110 \right\rangle, \left\vert 111 \right\rangle, $$ 

    
- 三量子比特展开式 $$ \left\vert \psi \right\rangle = \sum_{ijk=0,1} \alpha_{ijk} \left\vert ijk \right\rangle$$ 

### n 量子比特

- n 量子比特有$ 2^n $ 个计算基矢,构成正交归一完全集，记作：
  $$\{\left\vert i_1 i_2\cdots i_n \right\rangle\}$$
- n 量子比特展开式 $$ \begin{aligned}
  \left\vert \psi \right\rangle &= \sum_{i_1 i_2\cdots i_n =0,1} \alpha_{i_1 i_2\cdots i_n } \left\vert i_1 i_2\cdots i_n  \right\rangle \\
  &= \sum_{k=0} ^{2^n -1} a_{k} \left\vert k \right\rangle  
\end{aligned}$$ 

## Python 实验-1: Bloch球面

### 方法-1：
$$ \left\vert \psi \right\rangle = \alpha \left\vert 0 \right\rangle + \beta \left\vert 1 \right\rangle$$ 

```python
%matplotlib inline
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from qutip import *
from ipywidgets import interact, interactive, fixed

def plot_state1(alpha=1,beta=0,gamma=0):
    b = Bloch()
    spin_up = basis(2,0)
    spin_dn = basis(2,1)
    beta= complex(beta,gamma)
    tmp_state = (alpha*spin_up + beta*spin_dn).unit()
    b.add_states([tmp_state])
    b.render()
    return b

plot_state1(1,0,0)
```

```python
interact(plot_state1, alpha=(-1,1,0.1), beta=(-1,1,0.1), gamma = (-1,1,0.1));
```

### 方法-2 

\begin{aligned}
    \left\vert \psi \right\rangle =\cos\frac{\theta}{2}\left\vert 0 \right\rangle + e^{i\varphi} \sin\frac{\theta}{2}\left\vert 1 \right\rangle
  \end{aligned}

```python
%matplotlib inline
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from qutip import *
from ipywidgets import interact, interactive, fixed

def plot_state2(theta=0,varphi=0):
    b = Bloch()
    spin_up = basis(2,0)
    spin_dn = basis(2,1)
    varphi= complex(0,varphi)
    tmp_state = (np.cos(0.5*theta)*spin_up + np.exp(varphi)*np.sin(0.5*theta)*spin_dn).unit()
    b.add_states([tmp_state])
    b.render()
    return b

plot_state2(1,0)
```

```python
interact(plot_state2, theta=(0,np.pi+0.001,0.01), varphi=(0,2*np.pi+0.001,0.1));
```
